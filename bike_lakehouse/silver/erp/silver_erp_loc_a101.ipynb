# Initialization

```python
import pyspark.sql.functions as F
from pyspark.sql.types import StringType
from pyspark.sql.functions import trim, col
```

# Read Bronze Table

```python
df = spark.table("workspace.bronze.erp_loc_a101")
df.display()
```

# Silver Transformation

Dynamically trims all string columns in a DataFrame, rather than manually specifying each column name.

```python
for field in df.schema.fields:
    if isinstance(field.dataType, StringType):
        df = df.withColumn(field.name, trim(col(field.name)))
```

# Customer ID Cleaning

```python
df = df.withColumn("cid",F.regexp_replace(col("cid"),"-",""))
```

# Country Normalization

```python
df = df.withColumn(
    "cntry",
    F.when(col("cntry")=="DE","Germany")
      .when(col("cntry").isin("US","USA"),"United States")
      .when((col("cntry")=="") | col("cntry").isNull(),"n/a")
      .otherwise(col("cntry"))
)
```

# Rename Columns

```python
rename_map={
    "cid":"customer_number",
    "cntry":"country",
    }

for old_name,new_name in rename_map.items():
    df=df.withColumnRenamed(old_name,new_name)
```

```python
df.limit(10).display()
```

# Write Silver Table

```python
df.write.mode("overwrite").format("delta").saveAsTable("workspace.silver.erp_customer_location")
```
